```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure
```

```python
def ExtractObjects(image):
    img=np.copy(image)
    blob_labels=measure.label(img,background=0)
    number_of_objects=np.unique(blob_labels)
    return blob_labels,number_of_objects
```
