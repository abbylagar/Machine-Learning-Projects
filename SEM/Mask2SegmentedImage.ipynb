```python
import os                       # for working with files
import numpy as np              # for numerical computationss
import pandas as pd             # for working with dataframes
import matplotlib.pyplot as plt # for plotting informations on graph and images using tensors
from PIL import Image           # for checking images
import cv2
%matplotlib inline
```

```python
img_path ="/content/13-2.png"
mask_path = "/content/13-2 (2).png"

#load image
img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
```

```python
#multiply mask and image to get the segmented image
leaf_only = cv2.bitwise_and(img, img, mask=mask)
```

```python
#plot image
plt.imshow(leaf_only)
```

```python
#save image
cv2.imwrite(f"trial_image.png",leaf_only)
```
