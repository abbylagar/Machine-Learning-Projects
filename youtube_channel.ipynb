```python
for i in range(len(data)): 
  channel= ChannelsSearch(data['Name'][i], limit = 5, region = 'PH')
  channel = channel.result()
  if channel['result'] == []:
    print(data['Name'][i])
  else:  
    data['title'][i]=channel['result'][0]['title']
    data['id'][i]=channel['result'][0]['id']
    data['link'][i]=channel['result'][0]['link']
    data['subscribers'][i]=channel['result'][0]['subscribers']
```
